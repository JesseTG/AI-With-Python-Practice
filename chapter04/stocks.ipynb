# Finding Subgroups in Stock Market Data

**Affinity Propagation** is a clustering algorithm that doesn't require knowing the number of clusters in advance.

```python
import datetime
import json

import numpy
from matplotlib import pyplot
from sklearn import covariance, cluster
from mpl_finance import quotes_historical_yahoo_ochl as quotes_yahoo
```

```python
with open("company_symbol_mapping.json", 'r') as file:
    company_symbols = json.loads(file.read())

symbols, names = numpy.array(list(company_symbols.items())).T

start = datetime.datetime(2003, 7, 3)
end = datetime.datetime(2007, 5, 4)
quotes = tuple(quotes_yahoo(s, start, end, asobject=True) for s in symbols)

print(quotes)
```
